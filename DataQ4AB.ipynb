```python
data1 = {
    '활동': ['식도락 관광', '쇼핑', '자연경관 감상', '고궁/유적지 방문', '전통문화체험', '박물관 관람', 'K-POP 관련 공연장 방문', '드라마/영화 촬영지 방문',
           '공연 관람', '지역 축제', '유흥/오락', '뷰티/의료', '스포츠 관람/참가', '업무수행', '연수/교육/연구', '시설 시찰', '기타'],
    '인원수': [10893, 15508, 10557, 10047, 5828, 6753, 1906, 1425, 1058, 1426, 4760, 2522, 1650, 5915, 2968, 201, 308]
}

data2 = {
    '활동': ['식도락 관광', '쇼핑', '자연경관 감상', '고궁/유적지 방문', '전통문화체험', '박물관 관람', 'K-POP 관련 공연장 방문', '드라마/영화 촬영지 방문',
           '공연 관람', '지역 축제', '유흥/오락', '뷰티/의료', '스포츠 관람/참가', '업무수행', '연수/교육/연구', '시설 시찰', '기타'],
    '인원수': [8180, 10679, 7188, 6714, 3069, 2595, 1125, 590, 439, 484, 2345, 1150, 901, 4171, 924, 123, 288]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df1_sorted = df1.sort_values(by='인원수', ascending=False)
df2_sorted = df2.sort_values(by='인원수', ascending=False)

satisfaction_ratio = df2_sorted['인원수'] / df1_sorted['인원수']

data = {'활동': df1_sorted['활동'], '비율': satisfaction_ratio}

df_ratio = pd.DataFrame(data)
df_ratio_sorted = df_ratio.sort_values(by='비율', ascending=False)

plt.figure(figsize=(6, 6))

sns.heatmap(data=df_ratio_sorted[['비율']], annot=True, fmt=".3f", cmap='Purples', linewidths=0.5, xticklabels=1)

plt.yticks(range(len(df_ratio_sorted)), df_ratio_sorted['활동'], rotation=0)

title_font = {'fontsize': 14, 'fontweight': 'bold'}
plt.title('참여 인원수 대비 최고 만족도 활동의 비율', fontdict=title_font, loc='left')

plt.show()
```
