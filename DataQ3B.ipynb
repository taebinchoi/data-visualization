```python
data = {
    '고려요인': ['음식/미식', '쇼핑', '자연감상', '역사/유적', '패션 문화', 'K-POP 산업',
               '유흥/놀이', '치료/미용', '레저/스포츠', '여행시간', '이동거리', '경제적 비용', '기타'],
    '인원수': [6636, 5019, 3725, 3457, 1535, 1249, 1261, 322, 325, 1101, 296, 327, 1393]
}

df = pd.DataFrame(data)
df_sorted = df.sort_values(by='인원수', ascending=False)

colors = ['#8B0000', '#CD5C5C', '#CD5C5C', '#CD5C5C', '#CD5C5C', '#CD5C5C',
          '#CD5C5C', '#CD5C5C', '#CD5C5C', '#CD5C5C', '#CD5C5C', '#CD5C5C', '#CD5C5C']

plt.figure(figsize=(10, 6))

bar = plt.bar(df_sorted['고려요인'], df_sorted['인원수'], color=colors)

title_font = {'fontsize': 14, 'fontweight': 'bold'}
plt.title('방한 고려요인', fontdict=title_font, loc='left')
plt.xlabel('고려요인', size=13)

plt.legend(['인원수'], frameon=True, shadow=True, loc='upper right')

for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2.0, height, '%.0f' % height, ha='center', va='bottom', size=9)

plt.tick_params(axis='x', direction='in', length=5, pad=6, labelsize=11)
plt.tick_params(axis='y', direction='in', length=5)

plt.xticks(rotation=45, ha='right')

plt.show()
```
